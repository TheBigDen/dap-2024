## Вопросы и ответы

1. **Какие типы ячеек используются в Jupyter Notebook?**  
   В Jupyter Notebook используются два основных типа ячеек:  
   - **Кодовые ячейки (Code)** — для написания и выполнения программного кода.  
   - **Текстовые ячейки (Markdown)** — для написания описаний, инструкций и документации с использованием разметки Markdown.

2. **Как с помощью Markdown сформировать иерархический список?**  
   Для создания иерархического списка в Markdown используется символы `-` или `*`, с добавлением отступов для подуровней. Пример:  
   ```markdown
   - Первый уровень
     - Второй уровень
       - Третий уровень
3. **Как с помощью Markdown оформить цитату?**  
   Используется символ ">" перед текстом. Пример:
   > Это цитата.

4. **Как в текстовую ячейку вставить форматированный блок кода?**  
   Для форматированного блока кода в Markdown используются тройные обратные кавычки. Пример:
   ```python
   print("Hello, World!")
   ```

5. **Сколько уровней заголовка существует в Markdown?**  
   Markdown поддерживает шесть уровней заголовков. Они задаются символами "#":
   - # Заголовок 1 уровня
   - ## Заголовок 2 уровня
   - ### Заголовок 3 уровня и так далее.

8. **Что такое Jupyter Notebook, для чего используется?**  
   Jupyter Notebook — это интерактивная среда для программирования.  
   Она позволяет комбинировать код, текст и графики, что делает ее удобной для анализа данных, создания отчетов и обучения.

10. **Обязательно ли использовать Jupyter Notebook для анализа данных?**  
   Нет, Jupyter Notebook не является обязательным.  
   Анализ данных можно выполнять в других средах, таких как RStudio, PyCharm или скрипты Python.  
   Однако Jupyter Notebook удобен для визуализации и документирования анализа.
